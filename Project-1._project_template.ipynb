1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь
# Определяем функцию для преобразования столбца Образование и ВУЗ
def transform_education(edu_str):
    # Берем первые три слова и приводим к нижнему регистру
    first_words = ' '.join(edu_str.split()[:3]).lower()
    
    if 'высшее образование' in first_words:
        return 'высшее'
    elif 'неоконченное высшее' in first_words:
        return 'неоконченное высшее'
    elif 'среднее специальное' in first_words:
        return 'среднее специальное'
    elif 'среднее образование' in first_words:
        return 'среднее'
    else:
        return 'не указано'

# Создаем новый столбец
hh_database['Образование'] = hh_database['Образование и ВУЗ'].apply(transform_education)
count_secondary = (hh_database['Образование'] == 'среднее').sum()

print(f"Количество соискателей со средним образованием: {count_secondary}")
hh_database.drop('Образование и ВУЗ', axis=1, inplace=True)
```

* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
#ваш код здесь
def extract_gender(text):
    gender_part = text.split(' , ')[0].strip()  # Берём первую часть ("Мужчина" или "Женщина")
    return 'М' if gender_part == 'Мужчина' else 'Ж'

# Функция для извлечения возраста
def extract_age(text):
    age_part = text.split(' , ')[1].strip()  # Берём вторую часть ("39 лет" или "21 год")
    age = age_part.split()[0]  # Берём только число
    return int(age)

# Создаём новые столбцы
hh_database['Пол'] = hh_database['Пол, возраст'].apply(extract_gender)
hh_database['Возраст'] = hh_database['Пол, возраст'].apply(extract_age)

# Удаляем исходный столбец
hh_database.drop('Пол, возраст', axis=1, inplace=True)

# Проверяем первые 3 строки
print(hh_database[['Пол', 'Возраст']].head(3))

#Нахожу процент женских резюме
# 1. Количество женских резюме
female_count = (hh_database['Пол'] == 'Ж').sum()

# 2. Общее количество резюме
total_count = len(hh_database)

# 3. Вычисляем процент
female_percentage = (female_count / total_count) * 100

# 4. Округляем до сотых
female_percentage_rounded = round(female_percentage, 2)

print(f"Процент женских резюме: {female_percentage_rounded}%")

#Нахожу средний возраст соискателей
average_age = round(hh_database['Возраст'].mean(), 1)

print(f"Средний возраст соискателей: {average_age} лет")
```

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def get_experience(arg):
    if arg is np.nan or arg == 'Не указано':
        return None
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    arg_splitted = arg.split(' ')[:7]
    years = 0
    months = 0
    for index, item in enumerate (arg_splitted):
        if item in year_words:
            years = int(arg_splitted[index-1])
        if item in month_words:
            months = int(arg_splitted[index-1])
    return int(years*12 + months)

# Применяем исправленную функцию
hh_database['Опыт работы (месяц)'] = hh_database['Опыт работы'].apply(get_experience)

# Проверяем медиану
median_exp = hh_database['Опыт работы (месяц)'].median()
print(f"Медианный опыт работы: {round(median_exp)} месяцев")
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
# Список городов-миллионников
million_cities = [
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 
    'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 
    'Воронеж', 'Волгоград'
]

def parse_city_relocation_business_trips(row):
    if pd.isna(row):
        return {'Город': 'другие', 'Готовность к переезду': False, 'Готовность к командировкам': False}
    
    # Разбиваем строку по запятым, но сохраняем содержимое скобок
    parts = []
    current_part = []
    in_brackets = False
    for char in row:
        if char == '(':
            in_brackets = True
        elif char == ')':
            in_brackets = False
        if char == ',' and not in_brackets:
            parts.append(''.join(current_part).strip())
            current_part = []
        else:
            current_part.append(char)
    if current_part:
        parts.append(''.join(current_part).strip())
    
    # 1. Извлекаем город (первый элемент)
    city = parts[0].split(' , ')[0].strip()
    
    # Классифицируем город
    if city == 'Москва':
        city_category = 'Москва'
    elif city == 'Санкт-Петербург':
        city_category = 'Санкт-Петербург'
    elif city in million_cities:
        city_category = 'город-миллионник'
    else:
        city_category = 'другие'
    
    # 2. Готовность к переезду
    relocation = False
    for part in parts:
        part_lower = part.lower()
        if 'не готов к переезду' in part_lower or 'не готова к переезду' in part_lower:
            relocation = False
            break  # Явный отказ имеет приоритет
        elif ('готов к переезду' in part_lower or 
              'готова к переезду' in part_lower or 
              'хочу переехать' in part_lower):
            relocation = True
    
    # 3. Готовность к командировкам
    business_trips = False
    for part in parts:
        part_lower = part.lower()
        if 'не готов к командировкам' in part_lower or 'не готова к командировкам' in part_lower:
            business_trips = False
            break  # Явный отказ имеет приоритет
        elif ('готов к командировкам' in part_lower or 
              'готова к командировкам' in part_lower or 
              'редким командировкам' in part_lower or
              'возможны командировки' in part_lower):
            business_trips = True
    
    return {
        'Город': city_category,
        'Готовность к переезду': relocation,
        'Готовность к командировкам': business_trips
    }


# Применяем функцию и создаем новые столбцы
parsed_data = hh_database['Город, переезд, командировки'].apply(
    lambda x: parse_city_relocation_business_trips(x)
).apply(pd.Series)

# Добавляем новые столбцы в DataFrame
hh_database['Город'] = parsed_data['Город']
hh_database['Готовность к переезду'] = parsed_data['Готовность к переезду']
hh_database['Готовность к командировкам'] = parsed_data['Готовность к командировкам']

# Удаляем исходный столбец
hh_database.drop('Город, переезд, командировки', axis=1, inplace=True)

# Количество резюме из Санкт-Петербурга
spb_count = (hh_database['Город'] == 'Санкт-Петербург').sum()

# Общее количество резюме (исключая пропуски, если они есть)
total_count = hh_database['Город'].notna().sum()

# Вычисляем процент и округляем
spb_percentage = round((spb_count / total_count) * 100)

print(f"Процент соискателей из Санкт-Петербурга: {spb_percentage}%")

# 1. Создаём маску с явным указанием типа
mask = (
    hh_database['Готовность к переезду'].eq(True) & 
    hh_database['Готовность к командировкам'].eq(True)
)

# 2. Подсчитываем ТОЛЬКО строки, где оба значения не NA (даже если тип bool)
valid_rows = hh_database.loc[
    hh_database[['Готовность к переезду', 'Готовность к командировкам']].notna().all(axis=1)
]

ready_for_both = mask.sum()
total_valid = len(valid_rows)

# 3. Вычисляем процент
percentage = round((ready_for_both / total_valid) * 100) if total_valid > 0 else 0

print(f"Процент соискателей, готовых к переезду и командировкам: {percentage}%")
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
# Определяем все возможные категории для занятости и графика
employment_types = [
    'полная занятость', 
    'частичная занятость', 
    'проектная работа', 
    'волонтерство', 
    'стажировка'
]

schedule_types = [
    'полный день',
    'сменный график',
    'гибкий график',
    'удалённая работа',
    'вахтовый метод'
]

# Функция для создания бинарных признаков
def create_binary_features(df, column, categories):
    for category in categories:
        df[f'{column}_{category}'] = df[column].str.contains(category, na=False)
    return df

# Создаем бинарные признаки для занятости
hh_database = create_binary_features(hh_database, 'Занятость', employment_types)

# Создаем бинарные признаки для графика
hh_database = create_binary_features(hh_database, 'График', schedule_types)

# Удаляем исходные столбцы
hh_database = hh_database.drop(['Занятость', 'График'], axis=1)
```

```python
# 1. Находим пересечение (True в обоих столбцах)
project_and_volunteer = (
    hh_database['Занятость_проектная работа'] & 
    hh_database['Занятость_волонтерство']
).sum()

# 2. Общее количество резюме (для процента)
total = len(hh_database)

# 3. Вычисляем процент
percentage = round((project_and_volunteer / total) * 100, 1)

print(f"Количество соискателей, ищущих проектную работу и волонтерство: {project_and_volunteer}")
print(f"Процент от общего числа: {percentage}%")

# 1. Находим пересечение (True в обоих столбцах)
shift_and_flexible = (
    hh_database['График_вахтовый метод'] & 
    hh_database['График_гибкий график']
).sum()

# 2. Общее количество резюме (для процента)
total = len(hh_database)

# 3. Вычисляем процент
percentage = round((shift_and_flexible / total) * 100, 1)

print(f"Количество соискателей, готовых к вахтовому методу и гибкому графику: {shift_and_flexible}")
print(f"Процент от общего числа: {percentage}%")
```

В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:

Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
exchange_rates = pd.read_csv('data/ExchangeRates.csv')

# 1. Преобразование дат в обоих датафреймах
# Для hh_database (формат: день.месяц.год часы:минуты)
hh_database['Обновление резюме'] = pd.to_datetime(
    hh_database['Обновление резюме'], 
    format='%d.%m.%Y %H:%M'
).dt.date

# Для exchange_rates (пробуем оба возможных формата)
try:
    exchange_rates['date'] = pd.to_datetime(
        exchange_rates['date'], 
        format='%d.%m.%Y'
    ).dt.date
except ValueError:
    # Если первый формат не сработал, пробуем альтернативный (день/месяц/год)
    exchange_rates['date'] = pd.to_datetime(
        exchange_rates['date'], 
        format='%d/%m/%y'
    ).dt.date

# 2. Извлечение суммы и валюты из столбца ЗП
def extract_salary(salary_str):
    if pd.isna(salary_str):
        return None, None
    
    try:
        # Разделяем сумму и валюту
        parts = salary_str.split()
        amount = float(parts[0].replace(' ', ''))
        currency = parts[-1].replace('.', '')
        
        # Сопоставление валют с ISO кодами
        currency_map = {
            'руб': ('RUB', 1),
            'грн': ('UAH', 10),
            'USD': ('USD', 1),
            'EUR': ('EUR', 1),
            'белруб': ('BYN', 1),
            'KGS': ('KGS', 10),
            'сум': ('UZS', 10000),
            'AZN': ('AZN', 1),
            'KZT': ('KZT', 100)
        }
        
        iso_code, proportion = currency_map.get(currency, (None, None))
        return amount, iso_code
    except:
        return None, None

# Применяем функцию и создаем новые столбцы
hh_database[['ЗП_сумма', 'ЗП_валюта']] = hh_database['ЗП'].apply(
    lambda x: pd.Series(extract_salary(x))
)

# 3. Объединение с курсами валют
# Добавляем курс для рубля (1:1)
rub_rate = pd.DataFrame({
    'currency': ['RUB'],
    'date': [pd.to_datetime('01.01.2000').date()],  # Фиктивная дата
    'proportion': [1],
    'close': [1.0]
})
exchange_rates = pd.concat([exchange_rates, rub_rate])

# Объединяем данные
merged = pd.merge(
    hh_database,
    exchange_rates,
    left_on=['Обновление резюме', 'ЗП_валюта'],
    right_on=['date', 'currency'],
    how='left'
)

# 4. Расчет зарплаты в рублях
merged['ЗП (руб)'] = merged['ЗП_сумма'] * merged['close'] / merged['proportion']

# Заполняем пропуски:
# 1. Для зарплат в рублях
merged.loc[merged['ЗП_валюта'] == 'RUB', 'ЗП (руб)'] = merged['ЗП_сумма']
# 2. Для случаев, когда курс не найден (оставляем NaN)

# Переносим результат в исходный датафрейм
hh_database['ЗП (руб)'] = merged['ЗП (руб)']
```

```python
# Вычисляем медиану зарплаты в рублях (игнорируя пропуски)
median_salary_rub = hh_database['ЗП (руб)'].median()

# Переводим в тысячи рублей и округляем до целого
median_salary_k_rub = round(median_salary_rub / 1000)

print(f"Медианная желаемая зарплата: {median_salary_k_rub} тыс. руб.")

import pandas as pd

# Устанавливаем максимальное количество отображаемых столбцов
pd.set_option('display.max_columns', None)

# Теперь info() покажет все столбцы
hh_database.info()
```

7. Постройте сводную таблицу, иллюстрирующую зависимость **медианной** желаемой заработной платы от возраста (**"Возраст"**) и образования (**"Образование"**). На полученной сводной таблице постройте **тепловую карту**. Проанализируйте тепловую карту, сравнив показатели внутри групп.

```python
# ваш код здесь
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Создаем возрастные группы
age_bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 100]
age_labels = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60+']
hh_database['Возрастная группа'] = pd.cut(hh_database['Возраст'], bins=age_bins, labels=age_labels, right=False)

# Создаем сводную таблицу
pivot_table = hh_database.pivot_table(
    values='ЗП (руб)',
    index='Возрастная группа',
    columns='Образование',
    aggfunc='median',
    fill_value=0
)

# Переводим в тысячи рублей для удобства чтения
pivot_table_k = pivot_table / 1000

print("Сводная таблица (медианная зарплата в тысячах рублей):")
print(pivot_table_k.round(1))

# Строим тепловую карту
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table_k, 
            annot=True, 
            fmt='.1f', 
            cmap='YlOrRd', 
            linewidths=0.5,
            cbar_kws={'label': 'Медианная зарплата (тысячи рублей)'})

plt.title('Зависимость медианной заработной платы от возраста и образования', fontsize=16, fontweight='bold')
plt.xlabel('Образование', fontsize=12)
plt.ylabel('Возрастная группа', fontsize=12)
plt.tight_layout()
plt.show()

# Анализ карьерного роста (скорость роста зарплаты с возрастом)
growth_rates = {}

for education in pivot_table_k.columns:
    # Берем зарплаты для всех возрастных групп для данного образования
    salaries = pivot_table_k[education].values
    # Рассчитываем средний прирост между соседними возрастными группами
    growth = np.mean(np.diff(salaries) / salaries[:-1] * 100)  # в процентах
    growth_rates[education] = growth

# Сортируем по скорости роста
sorted_growth = sorted(growth_rates.items(), key=lambda x: x[1], reverse=True)

print("\nСкорость карьерного роста по образованию (средний % прироста между возрастными группами):")
for education, growth in sorted_growth:
    print(f"{education}: {growth:.1f}%")

# Определяем образование с самым быстрым ростом
fastest_growth_education, fastest_growth_rate = sorted_growth[0]
print(f"\nСамый быстрый карьерный рост наблюдается для образования: '{fastest_growth_education}'")
print(f"Средний ежегодный прирост: {fastest_growth_rate:.1f}% между возрастными группами")

# Дополнительный анализ: абсолютный прирост за всю карьеру
career_growth = {}
for education in pivot_table_k.columns:
    start_salary = pivot_table_k[education].iloc[0]  # первая возрастная группа
    peak_salary = pivot_table_k[education].max()     # пиковая зарплата
    growth = (peak_salary / start_salary - 1) * 100  # общий прирост в %
    career_growth[education] = growth

print("\nОбщий карьерный рост (от начала до пика):")
for education, growth in sorted(career_growth.items(), key=lambda x: x[1], reverse=True):
    print(f"{education}: +{growth:.1f}%")
```

Данные демонстрируют ярко выраженную зависимость между уровнем образования и динамикой карьерного роста. Высшее образование показывает наиболее впечатляющие результаты со средним приростом 7.2% между возрастными группами и общим ростом заработной платы на 100% от начала до пика карьеры. Это свидетельствует о том, что обладатели высшего образования получают максимальную отдачу от накопленного опыта и могут значительно увеличивать свой доход по мере профессионального развития.

Среднее специальное образование демонстрирует умеренный рост в 1.4% между возрастными группами с общим увеличением дохода на 50%, что указывает на более медленную, но стабильную динамику. Наиболее тревожная ситуация наблюдается у соискателей со средним образованием, которые показывают отрицательный рост (-15.3%) и минимальный общий прирост (+25%), что может быть связано с ограниченными возможностями для карьерного advancement и возрастной дискриминацией на рынке труда.
